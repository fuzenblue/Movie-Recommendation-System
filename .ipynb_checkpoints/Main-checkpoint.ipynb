Feature Selection Part

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'overview', 'genre']]
```

```python
movies
```

```python
movies['genre'].unique()
movies['genre'].head()
```

```python
movies = movies.copy()
movies['tags'] = movies['overview'] + movies['genre']
```

```python
movies
```

```python
new_data = movies.drop(columns=['overview'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title'] == "The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])

for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies, top_n=5):
    index = new_data[new_data['title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])

    results = []
    for i in distance[:top_n]:
        results.append(new_data.iloc[i[0]].title)
    return results
```

```python
recommand("Iron Man")
```

```python
non_string_genres = new_data[~new_data['genre'].apply(lambda x: isinstance(x, str))]
len(non_string_genres)
```

```python
new_data['genre'] = new_data['genre'].fillna('')
```

```python
new_data['genre'] = new_data['genre'].astype(str)
```

```python
new_data['genre'] = new_data['genre'].str.strip().str.replace(' ', '')
```

```python
new_data['genre'] = new_data['genre'].replace('nan', '')
```

```python
new_data['genre'] = new_data['genre'].apply(lambda x: x.split(',') if x != '' else [])
```

```python
new_data['genre'].head()
```

```python
empty_genre_entries = new_data[new_data['genre'].apply(lambda x: len(x) == 0)]
len(empty_genre_entries)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()
```

```python
genre_encoded = pd.DataFrame(mlb.fit_transform(new_data['genre']), columns=mlb.classes_, index=new_data.index)
```

```python
new_data = pd.concat([new_data, genre_encoded], axis=1)
```

### Select if have at least 1 genre ###

```python
def recommend_by_genres_or(genres_to_search):
    filtered_movies = new_data[new_data['genre'].apply(lambda x: any(genre in x for genre in genres_to_search))]
    
    if not filtered_movies.empty:
        top_movies = filtered_movies.head(5)
        for _, row in top_movies.iterrows():
            print(row['title'])
    else:
        print(f"No movies found with genres: {genres_to_search}")
```

```python
genres_to_search = ["War"]
recommend_by_genres_or(genres_to_search)
```

### Select if have all genre ###

```python
def recommend_by_genres(genres, top_n=5):
    filtered_data = new_data[new_data['genre'].apply(lambda x: all(g in x for g in genres))]

    if not filtered_data.empty:
        for i, title in enumerate(filtered_data['title'].head(top_n), start=1):
            print(f"{i}. {title}")
    else:
        print(f"No movies found with genres: {genres}")
```

```python
# genres_to_search = ["War"]
genres_to_search = ["Drama", "War", "Music"]
recommend_by_genres(genres_to_search, top_n=5)
```

## Recommend By Genres & Title ##

```python
def recommend_by_genres_and_title(movies, genres, top_n=5):

    movie_based_recommendations = recommand(movies, top_n=len(new_data))
    filtered_data = new_data[new_data['genre'].apply(lambda x: all(g in x for g in genres))]

    if not filtered_data.empty:
        genre_counts = {}
        for movie_title in movie_based_recommendations:
            movie_row = filtered_data[filtered_data['title'] == movie_title]
            if not movie_row.empty:
                genres_in_movie = movie_row.iloc[0]['genre']
                count = sum([1 for g in genres if g in genres_in_movie])
                genre_counts[movie_title] = count

        sorted_movies = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)
        results = [movie[0] for movie in sorted_movies[:top_n]]
        return results
    else:
        print(f"No movies found with genres: {genres}")
        return []
```

```python
movies = "Joker"
genres = ["Drama", "Music"]

result = recommend_by_genres_and_title(movies, genres)
print("Movies recommended by title and genres:")
print(result)
```
